```python
import numpy
from syntheticyeastcells import create_samples, colored_segmentation_map
```

```python
settings = {
  'n_cells_per_image': [10, 100, 1000],
  'min_distance_boundary': [0, 50, 150],
  'r0_range': [(2, 5), (5, 20), (20, 40)],
  'r1_factor_range': [(0.7, 0.8), (1.0, 1.0)],
  'spatial_blur_std': [0.5, 1.5, 3],
  'background_intensity': [0.1, 0.4, 0.8],
  'background_contrast': [0.001, 0.002, 0.01],
  'core_contrast': [0.01, 0.02, 0.1],
  'p_white_outside': [0, 0.5, 1],
}
```

```python
from matplotlib import pyplot
%matplotlib inline

pyplot.rcParams['font.size']=24
pyplot.rcParams['font.family']='DejaVu Serif'

for name, values in settings.items():
  
  fig, axes = pyplot.subplots(1, len(values), figsize=(10, 10/len(values)))
  axes = axes.ravel()
  for axis, value in zip(axes, values):
    images, labels = create_samples(n_images=1, **{name: value})
    labels = colored_segmentation_map(labels)
    fig.suptitle(f'{name}')
    axis.imshow(images[0])
    axis.imshow(labels[0])
    axis.set_xticks([]); axis.set_yticks([])
    axis.set_xlabel(f'{value}')
  pyplot.tight_layout()
#   fig.savefig(f'/home/p253591/repositories/synthetic-yeast-cells/images/example-settings-{name}.png')
```
